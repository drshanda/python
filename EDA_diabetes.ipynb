```python
df = pd.read_csv("/Users/lashandawilliams/python/diabetes.csv")
```

```python
df.drop_duplicates(inplace = True)
```

```python
df = df[df["Glucose"] > 0]
df = df[df["BloodPressure"] > 0]
df = df[df["Insulin"] > 0]
df = df[df["BMI"] > 0]
```

```python
pregnancies = df['Pregnancies'].value_counts()

fig, ax = plt.subplots(figsize = [7,7])

ax.bar(pregnancies.index, pregnancies.values)
ax.set_xlim([0, 17])


```
