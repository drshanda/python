```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from itertools import combinations

%matplotlib inline
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Iris.csv")
```

```python
df.shape
```

```python
df.describe()
```

```python
df.drop("Id", axis = 1, inplace = True)
```

```python

colors = ['blue', 'darkorange', 'green']

fig, ax = plt.subplots()

ax.bar(df.Species.value_counts().index, df.Species.value_counts().values, color = colors)
ax.set_xlabel("Species")
ax.set_ylabel("Count")

```

```python
column_names = df.drop("Species", axis = 1).columns

for i in column_names:
  fig, ax = plt.subplots()
  sns.boxplot(x = "Species", y = i, hue = "Species", data = df)
  plt.xlabel(i)
```

```python
for i in column_names:
  fig, ax = plt.subplots()
  sns.histplot(x = i, kde=True, data = df, hue = "Species")
```

```python
color_map = {'Iris-setosa': 'blue', 'Iris-versicolor': 'orange', 'Iris-virginica': 'green'}

# Create a list of colors based on the categories
point_colors = [color_map[cat] for cat in df["Species"]]

for col1, col2 in combinations(df.drop("Species", axis = 1).columns, 2): # <-----
    plt.scatter(df[col1], df[col2], c = point_colors)
    plt.xlabel(col1)
    plt.ylabel(col2)
    plt.show()
```

```python
sns.pairplot(df, hue = "Species")
```

```python
sns.heatmap(df.drop(["Species"], axis = 1).corr(),
            annot=True,
            cmap='coolwarm',
            fmt=".2f")
```
